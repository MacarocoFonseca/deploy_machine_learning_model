Now test on the validation data set but first I will add everything on function

```python
def read_dataframe(filename):
    '''Data preprocessing
    '''
    df = pd.read_parquet(filename)

    df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)

    df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds()/60)

    df = df[(df.duration >= 1) & (df.duration <= 60)]

    categorical = ['PULocationID','DOLocationID']
    df[categorical] = df[categorical].astype(str)

    return df
```

```python
df_train = read_dataframe('green_tripdata_2023-01.parquet')
df_val = read_dataframe('green_tripdata_2023-02.parquet')
```

```python
len(df_train),len(df_val)
```

```python
df_train['PU_DO'] = df_train['PULocationID'] + '_' + df_train['DOLocationID']
df_val['PU_DO'] = df_val['PULocationID'] + '_' + df_val['DOLocationID']
```

```python
categorical = ['PU_DO']#'PULocationID','DOLocationID']
numerical = ['trip_distance'] 

dv = DictVectorizer()

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
# target
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

mean_squared_error(y_val, y_pred, squared=False)
```

```python
lr = Lasso(alpha=0.001)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

mean_squared_error(y_val, y_pred, squared=False)
```

```python
lr = Ridge(alpha=0.001)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

mean_squared_error(y_val, y_pred, squared=False)
```

```python
sns.distplot(y_pred, label='prediction');
sns.distplot(y_train, label='actual');

plt.legend();
```

### This is our baseline, lets try to improve it

```python
# Lasso regression, has regularization

lr = Lasso() 
lr.fit(X_train,y_train)

y_pred = lr.predict(X_train)

sns.distplot(y_pred, label='prediction');
sns.distplot(y_train, label='actual');

plt.legend();
```
